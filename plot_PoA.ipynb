```python
import pandas as pd
import matplotlib.pyplot as plt

# read data from CSV file
df = pd.read_csv(data_file_name)

# filter rows based on instance name
md_cust = df[df['Instance'].str.startswith('MD_cust')]
mdtw_cust = df[df['Instance'].str.startswith('MDTW_cust')]

# plot the data with different line styles
plt.plot(nb_customer, md_cust['PoA_U'], label='Utility-based without TW', linestyle='--')
plt.plot(nb_customer, md_cust['PoA_C'], label='Cost-based without TW', linestyle='--')
plt.plot(nb_customer, mdtw_cust['PoA_U'], label='Utility-based with TW', linestyle=':')
plt.plot(nb_customer, mdtw_cust['PoA_C'], label='Cost-based with TW', linestyle='--')
plt.xlabel('Number of Customers')
plt.ylim((0.9,3.5))
plt.ylabel('Price of Anarchy')
plt.legend()
plt.savefig('PoA.pdf')
plt.show()
```
